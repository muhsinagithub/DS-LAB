```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [2,4,1]
plt.plot(x,y,linewidth=6)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('my first graph!')
plt.show()
```

```python
from matplotlib import pyplot as plot
y=[10,5,8,4,2,2,2,5]
plt.hist(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[2,4,1]
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('my first graph1')
plt.show()
```

```python
import plotly.express as px
fig=px.line(x=[1,2,3],y=[1,2,3])
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
print(df)
fig=px.line(df,x="species",y="petal_width")
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.bar(df,x="sepal_width",y="sepal_length")
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.histogram(df,x="sepal_length",y="petal_length")
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter(df,x="species",y="petal_width")
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter(df,x="species",y="petal_width",size="petal_length",color="species")
fig.show()
```

```python
import plotly.express as px
df=px.data.iris()
fig=px.scatter_3d(df,x='sepal_width',y='sepal_length',z='petal_width',color='species')
fig.show()
```
